```python
from sklearn.datasets import load_iris
```

```python
iris= load_iris()
```

```python
x=iris.data
```

```python
y=iris.target
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(x,y, test_size=0.2)
```

```python
X_train.shape
```

```python
y_train.shape
```

```python
X_test.shape
```

```python
y_test.shape
```

```python
# apply algorthim
from sklearn.tree import DecisionTreeClassifier
clf= DecisionTreeClassifier()
```

```python
# Train our model
clf.fit(X_train,y_train)
```

```python
y_pred = clf.predict(X_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test,y_pred)
```

```python
from sklearn.tree import plot_tree
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 80,50
plot_tree(clf)
```

Social network ads - Decision trees

```python
import pandas as pd
import numpy as np
```

```python
ads_dataset = pd.read_csv("/content/Social_Network_Ads.csv")
```

```python
ads_dataset.head()
```

```python
# since gender is categorical change into numerical
ads_dataset['Gender'].replace({'Male':0,'Female':1},inplace=True)
```

```python
X= ads_dataset.iloc[:,1:4].values
y= ads_dataset.iloc[:,-1].values
```

```python
X.shape
```

```python
y.shape
```

```python
# classifier

clf1 = DecisionTreeClassifier()
```

```python
clf1.fit(X,y)
```

```python
from matplotlib.pylab import rcParams
rcParams['figure.figsize'] = 80,50
plot_tree(clf1)
```

```python
# overfitting  assign max depth as 3
clf1 = DecisionTreeClassifier(max_depth=3)
```

```python
clf1.fit(X,y)
```

```python
rcParams['figure.figsize'] = 80,50
plot_tree(clf1)
```
