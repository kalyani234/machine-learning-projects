```python
import pandas as pd
import seaborn as sns
from matplotlib import pyplot as plt
import numpy as np
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
df = pd.DataFrame(iris.data, columns = iris.feature_names)
df.head()
```

```python
iris.target
```

```python
iris.target_names
```

```python
df['target'] =  iris.target
df.head()
```

```python
df[df.target==1].head()
```

```python
df['flower_name'] =  df.target.apply(lambda x : iris.target_names[x])
df.head()
```

```python
# Viz
```

```python
df0 = df[df.target==0]
df1 = df[df.target==1]
df2 = df[df.target==2]
```

```python
plt.xlabel("sepal length (cm)")
plt.ylabel("sepal width (cm)")
plt.scatter(df0["sepal length (cm)"],df0["sepal width (cm)"],marker ="*", color = "red")
plt.scatter(df1["sepal length (cm)"],df1["sepal width (cm)"],marker =".", color = "blue")
```

```python
plt.xlabel("petal length (cm)")
plt.ylabel("petal width (cm)")
plt.scatter(df0["petal length (cm)"],df0["petal width (cm)"],marker ="*", color = "red")
plt.scatter(df1["petal length (cm)"],df1["petal width (cm)"],marker =".", color = "blue")
```

```python
from sklearn.model_selection import train_test_split 
```

```python
X = df.drop(["target","flower_name"],axis="columns")
```

```python
X
```

```python
y = df.target
y
```

```python
 X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.33)
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
y_test
```

```python
from sklearn.svm import SVC
```

```python
sv = SVC()
```

```python
sv.fit(X_train,y_train)
```

```python
sv.score(X_test,y_test)
```

```python
sv.predict(pd.DataFrame([[6.0,3.4,4.5,1.6]],columns = ["sepal length (cm)","sepal width (cm)","petal length (cm)","petal width (cm)"]))
```

```python
digits_df = pd.DataFrame(digits.data, columns = digits.feature_names)
digits_df.head(1)
```

```python
digits_df['target'] = digits.target
digits_df
```

```python
digits_df['digit_name'] =  digits_df.target.apply(lambda x : digits.target_names[x])
digits_df
```

```python
from sklearn.model_selection import train_test_split
```

```python
X = digits_df.drop(["target","digit_name"], axis ="columns")
X[:1]
```

```python
y = digits_df.target
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
len(X_train)
```

```python
len(X_test)
```

```python
from sklearn.svm import SVC
```

```python
sv1 = SVC(kernel='linear')
```

```python
sv1.fit(X_train,y_train)
```

```python
sv1.score(X_test,y_test)
```

```python
pred = sv1.predict(X_test)
```
